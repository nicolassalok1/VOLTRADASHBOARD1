```python

spot_slider = widgets.FloatSlider(value=1.0, min=0.5, max=1.5, step=0.1, description='S_0')
strike_slider = widgets.FloatSlider(value=1.0, min=0.5, max=1.5, step=0.1, description='Strike')

sigma_slider = widgets.FloatSlider(value=0.2, min=0.01, max=1.0, step=0.01, description='Sigma')
r_slider = widgets.FloatSlider(value=0.02, min=-0.05, max=0.1, step=0.005, description='r')
T_slider = widgets.FloatSlider(value=1.0, min=0.05, max=2.0, step=0.05, description='T')

output = widgets.Output()

def _update(change=None):
    with output:
        output.clear_output()
        s0 = spot_slider.value
        k = strike_slider.value
        premium = 0.0
        payoff_grid = np.array([s0])
        s_grid = np.array([s0])
        pnl_grid = payoff_grid - premium
        fig, ax = plt.subplots()
        ax.plot(s_grid, payoff_grid)
        plt.show()
        plt.close(fig)

for sl in (spot_slider, strike_slider, sigma_slider, r_slider, T_slider):
    sl.observe(_update, names='value')
_update()
display(widgets.VBox([spot_slider, strike_slider, sigma_slider, r_slider, T_slider, output]))
```
